```python
# importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import keras
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout
from tensorflow.keras.layers import GlobalMaxPooling2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers, optimizers
from tensorflow.keras.utils import to_categorical
```

```python
# import acc+score and cifar10 predefined dataset from tensorflow keras dataset
from sklearn.metrics import accuracy_score
from tensorflow.keras.datasets import cifar10
```

```python
# splitting the train and test data from the cifar10 dataset and [0] shows the 1sr=t image in cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
plt.imshow(x_train[0])
```

```python
# shows the 6th image of cifar10
plt.imshow(x_train[5])
```

```python
# tis is not an image so plt is not used here
x_train[1]
```

```python
# normalizing the pic values btw 0 and 1 & 0-black , 255-white. cifar10 pic values are represented as an int btw 0 and 255
x_train = x_train/255
x_test = x_test/255
```

```python
#performs one hot encoding i.e categorical to numerical & here y_train,y_test class labels are converted yo numerical
y_train_en = to_categorical(y_train, 10)
y_test_en = to_categorical(y_test, 10)
```

```python
# linear stack for model
model = Sequential()
# filter 32 filter size(4,4) image pixel (32,32) with 3 colors(rgb) relu-introduce non linearity to the model
model.add(Conv2D(32,(4,4), input_shape = (32,32,3),activation = 'relu'))
# maxpooling-reduces spatial dimensions (all edges and corners) with pooling window size (2,2)
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Conv2D(32,(4,4), input_shape = (32,32,3),activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))

#converts multidimension into iD vector
model.add(Flatten())
# provides fully connected layer with active fnc = relu
model.add(Dense(128,activation = 'relu'))
model.add(Dense(10,activation = 'relu'))
# compiles and loss - multiclass classification
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam', metrics=['accuracy'])
```

```python
# summary of the model
model.summary()
```

```python
# training the model with 20 epochs(passing dataset) verbose-amount of output display and validation provides validation during training of the model
history = model.fit(x_train, y_train_en, epochs = 20, verbose = 1,validation_data=(x_test,y_test_en))
```
