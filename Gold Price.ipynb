**Gold Price Prediction**


```python
# importing the necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# loading dataset
df = pd.read_csv("/content/gld_price_data.csv")
df.head()
```

```python
# splitting x and y
x = df[['SPX','USO','SLV','EUR/USD']]
y = df['GLD']
```

```python
# splitting train and test data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
scaler = MinMaxScaler()
# scaling is normalizing the values to mean = 0 and variance = 1
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
# linear stack and fully covered layers with activation fnc
model = Sequential()
model.add(Dense(10,activation = 'relu',input_dim=4))
model.add(Dense(10,activation = 'relu'))
model.add(Dense(1,activation = 'linear'))
model.compile(loss = "mean_squared_error",optimizer = 'Adam')
```

```python
# iterating through the training data and splitting the 10% of validation set
history = model.fit(x_train_scaled, y_train, epochs=50,validation_split = 0.1)
```

```python
# predicting the x_test_scaled
y_pred = model.predict(x_test_scaled)
```

```python
# accuracy score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
